```python
import torch
import numpy as np
import pandas as pd
import matplotlib
print("PyTorch version:", torch.__version__)
print("NumPy version:", np.__version__)
print("Pandas version:", pd.__version__)
print("Matplotlib version:", matplotlib.__version__)
print("All packages loaded successfully.")
```

```python
#df = pd.read_csv("data/DSL-StrongPasswordData.csv")
#print("Shape:", df.shape)
#print("\nColumn names:")
#print(df.columns.tolist())
#print("\nFirst row:")
#print(df.iloc[0])

df = pd.read_csv("data/DSL-StrongPasswordData.csv")

print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns\n")

print("Column names:")
for col in df.columns:
    print(f"  - {col}")

print("\nFirst row (as a table):")
df.head(1)
```

```python
print(df['subject'].nunique())
print(df['subject'].value_counts())
```

```python
feature_cols = [col for col in df.columns if col not in ['subject', 'sessionIndex', 'rep']]
print(f"Number of feature columns: {len(feature_cols)}")
print(feature_cols)
```

```python
# Separate feature columns from metadata columns
metadata_cols = ['subject', 'sessionIndex', 'rep']
feature_cols = [col for col in df.columns if col not in metadata_cols]

print(f"Total columns: {len(df.columns)}")
print(f"Metadata columns: {len(metadata_cols)} -> {metadata_cols}")
print(f"Feature columns: {len(feature_cols)}\n")

print("Feature columns in order:")
for i, col in enumerate(feature_cols, start=1):
    print(f"  {i:2d}. {col}")
```

```python
import numpy as np

def extract_sequence(row, feature_cols):
    """
    Converts one row of the dataframe into a (11, 3) array.
    Each of the first 10 timesteps holds [H, DD, UD] for one key transition.
    The 11th timestep holds [H.Return, 0.0, 0.0] since there is no transition after it.
    """
    values = row[feature_cols].values.astype(np.float32)
    
    # First 30 values form 10 complete triplets of (H, DD, UD)
    full_transitions = values[:30].reshape(10, 3)
    
    # Last value is H.Return, pad with zeros for DD and UD
    last_timestep = np.array([values[30], 0.0, 0.0], dtype=np.float32)
    
    sequence = np.vstack([full_transitions, last_timestep])
    return sequence

# Test it on the first row
test_sequence = extract_sequence(df.iloc[0], feature_cols)
print(f"Shape: {test_sequence.shape}")
print(f"\nSequence:\n{test_sequence}")
```

```python
# Apply the function to every row in the dataframe
all_sequences = np.array([extract_sequence(row, feature_cols) for _, row in df.iterrows()])

print(f"Full dataset sequence array shape: {all_sequences.shape}")
print(f"Expected shape: (20400, 11, 3)")
print(f"Match: {all_sequences.shape == (20400, 11, 3)}")

print(f"\nAny NaN values present: {np.isnan(all_sequences).any()}")
print(f"Min value in dataset: {all_sequences.min()}")
print(f"Max value in dataset: {all_sequences.max()}")
```

```python
# Check the distribution of timing values
flat_values = all_sequences.reshape(-1)

print(f"Total values: {len(flat_values)}")
print(f"\nPercentile breakdown:")
percentiles = [1, 5, 25, 50, 75, 95, 99, 99.9]
for p in percentiles:
    print(f"  {p:5.1f}th percentile: {np.percentile(flat_values, p):.4f}s")

print(f"\nValues above 2.0 seconds: {(flat_values > 2.0).sum()} ({(flat_values > 2.0).mean()*100:.2f}%)")
print(f"Values above 5.0 seconds: {(flat_values > 5.0).sum()} ({(flat_values > 5.0).mean()*100:.2f}%)")
```

```python
# Clip outliers at 2.0 seconds
all_sequences_clipped = np.clip(all_sequences, a_min=None, a_max=2.0)

print(f"Before clipping - Max value: {all_sequences.max():.4f}s")
print(f"After clipping  - Max value: {all_sequences_clipped.max():.4f}s")
print(f"Min value unchanged: {all_sequences_clipped.min():.4f}s")

# Now apply normalization (z-score per feature across the full dataset)
mean = all_sequences_clipped.mean(axis=(0,1), keepdims=True)
std  = all_sequences_clipped.std(axis=(0,1), keepdims=True)

all_sequences_normalized = (all_sequences_clipped - mean) / (std + 1e-8)

print(f"\nAfter normalization:")
print(f"  Mean: {all_sequences_normalized.mean():.6f} (should be close to 0)")
print(f"  Std:  {all_sequences_normalized.std():.6f} (should be close to 1)")
print(f"  Shape: {all_sequences_normalized.shape}")

# Save mean and std for later use during inference
print(f"\nFeature mean (per feature): {mean.squeeze()}")
print(f"Feature std  (per feature): {std.squeeze()}")
```

```python
import os

# Create processed data directory
os.makedirs("data/processed", exist_ok=True)

# Save the normalized sequences
np.save("data/processed/sequences_normalized.npy", all_sequences_normalized)

# Save labels (subject IDs) aligned to each sequence
labels = df['subject'].values
np.save("data/processed/labels.npy", labels)

# Save normalization parameters for later use during inference
np.save("data/processed/feature_mean.npy", mean)
np.save("data/processed/feature_std.npy", std)

# Verify saves
print("Saved files:")
for f in os.listdir("data/processed"):
    size = os.path.getsize(f"data/processed/{f}")
    print(f"  {f}: {size:,} bytes")

print(f"\nSequences shape: {all_sequences_normalized.shape}")
print(f"Labels shape: {labels.shape}")
print(f"Unique subjects: {len(np.unique(labels))}")
print(f"\nSample label alignment check:")
print(f"  First label: {labels[0]}, Last label: {labels[-1]}")
```
